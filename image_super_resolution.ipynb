```
from tqdm import *
import numpy as np
from PIL import Image
from ISR.models import RDN
import imageio

rdn = RDN(arch_params={'C':6, 'D':20, 'G':64, 'G0':64, 'x':2})
rdn.model.load_weights('image-super-resolution/weights/rdn-C6-D20-G64-G064-x2_ArtefactCancelling_epoch219.hdf5')

for mclass in ['m_glass', 'm_nglass']:#['f_glass', 'f_nglass', 'm_glass', 'm_nglass']:

    if not os.path.exists('image-super-resolution/data/%s'%mclass):
        os.makedirs('image-super-resolution/data/%s'%mclass)

    from os import walk

    images = []
    for (dirpath, dirnames, filenames) in walk('data/%s'%mclass):
        for filename in tqdm(filenames):
            if '.png' not in filename:
                continue
            img = Image.open('data/%s/%s'%(mclass, filename))
            lr_img = np.array(img)

            sr_img = rdn.predict(lr_img)
            imageio.imwrite('image-super-resolution/data/%s/%s'%(mclass, filename), sr_img)
```
